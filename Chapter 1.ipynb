```python
class Man:
	def __init__ (self,name):
		self.name = name
		print("initialized!")
	
	def hello(self):
		print("Hello" + self.name + "!")

	def goodbye(self):
		print("goodbye" + self.name + "!")

m = Man("Tom")
m.hello()
m.goodbye()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,6,0.1) #以0.1为单位，生成0-6的数据
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.show
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1,label = "sin")
plt.plot(x,y2,linestyle = "--", label = "cos") # 用虚线绘制
plt.xlabel("x") #x轴标签
plt.ylabel("y") # y轴标签
plt.title("sin&cos") #标题
plt.legend()
plt.show()
```
