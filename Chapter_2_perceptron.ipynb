• 感知机是具有输入和输出的算法。给定一个输入后，将输出一个既定的值。
• 感知机将权重和偏置设定为参数。
• 使用感知机可以表示与门和或门等逻辑电路。
• 异或门无法通过单层感知机来表示。
• 使用2层感知机可以表示异或门。
• 单层感知机只能表示线性空间，而多层感知机可以表示非线性空间。
• 多层感知机（在理论上）可以表示计算机。

```python
def temperature(x):
    if x > 35:
        print("太热了！")
    elif x > 25:
        print("天气不错~")
    elif x > 15:
        print("有点凉")
    else:
        print("太冷了！")
```

```python
temperature(36)
```

AND 函数  与门，输入同时为1输出1，否则输出0

```python
def AND(x1,x2):
    w1,w2,theta = 0.5,0.5,0.7
    tmp = x1*w1 + x2*w2
    if tmp <=theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
AND(0,0)
```

```python
AND(1,1)
```

```python
def AND(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
def NAND(x1,x2):
    x = np.array([x1,x2])
    w = np.array([-0.5,-0.5])
    b = 0.7
    tmp = np.sum(w*x)+b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
def OR(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5,0.5])
    b = -0.3
    tmp = np.sum(w*x)+b
    if tmp <= 0:
        return 0
    else:
        return 1
```

单层感知机无法表示分线性空间，因此无法表示异或门，引入叠加层(multi-layered perceptron)表示非线性空间，通过组合与门、与非门、或门实现异或门

```python
def XOR (x1,x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)
    y = AND(s1,s2)
    return y
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-2, 2, 400)
sigmoid = 1 / (1 + np.exp(-x))
tanh = np.tanh(x)
relu = np.maximum(0, x)
plt.figure(figsize=(8, 6))
plt.plot(x, sigmoid, label='Sigmoid', color='blue', linewidth=1)
plt.plot(x, tanh, label='tanh', color='green', linewidth=1)
plt.plot(x, relu, label='ReLU', color='red', linewidth=1)
plt.title("Activation Functions: Sigmoid vs tanh vs ReLU")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
